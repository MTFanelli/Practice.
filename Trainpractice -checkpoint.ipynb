#Numerical data distribution

list(set(df.dtypes.tolist()))



```python
df_num = df.select_dtypes(include = ['float64', 'int64'])
df_num.head()
```

```python
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8); # ; avoid having the matplotlib verbose informations
```

```python
#correlation

df_num_corr = df_num.corr()['ApplicantIncome'][:-1] # -1 because the latest row is Applicant Income
golden_features_list = df_num_corr[abs(df_num_corr) > 0.5].sort_values(ascending=False)
print("There is {} strongly correlated values with ApplicantIncome:\n{}".format(len(golden_features_list), golden_features_list))
```

```python
#Outliers

for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num,
                x_vars=df_num.columns[i:i+5],
                y_vars=['ApplicantIncome'])
```
